# initialize the WCS coordinates 

one of the purposes of this exercise is to test how big of an effect taking into account the center of mass fluctuation from realization to realization would be 

```python
hdulist = astropy.io.fits.open(fits)
prihdr = hdulist[0].header
w = wcs.WCS(prihdr)
```

```python
NW_pix = w.wcs_world2pix(
    np.array([inputRDF["NW_ra"], inputRDF["NW_dec"]]).transpose(), 1)
NW_pix = NW_pix.transpose()
```

```python
SE_pix = w.wcs_world2pix(
    np.array([inputRDF["SE_ra"], inputRDF["SE_dec"]]).transpose(), 1)
SE_pix = SE_pix.transpose()
```

sanity check! we should be able to recover d_proj with the pairs of wcs coordinates! 

```python
#%psource angular_separation
```

check against http://hea.iki.rssi.ru/AZT22/ENG/cgi-bin/c_dist.htm

```python
cal_ang_sep = angular_separation(inputRDF["NW_ra"] / 180. * np.pi, 
                               inputRDF["NW_dec"] / 180. * np.pi, 
                               inputRDF["SE_ra"] / 180. * np.pi, 
                               inputRDF["SE_dec"] / 180. * np.pi)
```

The above gives radian as the output

```python
d_proj = cosmology.angular_diameter_distance(z=0.87) * cal_ang_sep
d_proj = d_proj.value
```

This following test should pass if the d_projs are calculated properly

```python
assert np.allclose(d_proj, data["d_proj"]), \
    "d_proj not calculated properly"
```

# Calculate the center of mass

$\vec{x}_{CM} = (\vec{x}_{NW}\times m_{NW} + \vec{x}_{SE}\times m_{SE})/m_{total}$

```python
CM_pix = [(NW_pix[i] * data["m_1"] + SE_pix[i] * data["m_2"]) / \
     (data["m_1"] + data["m_2"]) for i in range(2)]
CM_pix = np.array(CM_pix)
print "proper CM pix coords are {0}, {1}".format(
    CM_pix[0].mean(), CM_pix[1].mean())
CM_wcs = w.wcs_pix2world([[CM_pix[0].mean(), CM_pix[1].mean()], 
                          [CM_pix[0].mean(), CM_pix[1].mean()]], 1)
CM_wcs = CM_wcs[0]
```

### calculate how long 1 kpc should correspond to in the plot

```python
cosmology.angular_diameter_distance(z=0.87).value
```

```python
cosmology = FlatLambdaCDM(H0=70, Om0=0.3)

# this tells us how many Mpc there is in 1 radian -> Mpc in 1 arcsec 
MpcIn1Arcsec = \
    cosmology.angular_diameter_distance(z=0.87).value * \
    np.pi / 180. / 60. / 60.
```

we want to know how many arcsecs 100 kpcs correspond to  

```python
ArcSecIn100kpc = 1 / MpcIn1Arcsec * 0.1  # convert from 1 Mpc to kpc
ArcSecIn100kpc
```

This estimate is off from James' estimate by 2 arcsecs for god-knows-why

```python
ArcSecIn100kpc / 60 / 60. + 15.744642
```

now this is the hypothenuese of our right angled triangle,
we need to figure out the angle before we can find the corresponding differences in RA and Dec. From previous effort to figure out the CD matrix, we know the mean rotation angle is 

```python
rotAng = 42.592869805726998 / 180. * np.pi # degree to radian
# convert from 1 normal degree to 1 RA degree
delRA = ArcSecIn100kpc * np.sin(rotAng) / 15.
delDec = ArcSecIn100kpc * np.cos(rotAng)
```
