```python
import h5py 
from MCMAC import vrad
from astroML.plotting import hist
import pickle
```

```python
NW_z = pickle.load(open("../call_MCMAC/NW_bootstrapRedshift.pickle", "r"))
SE_z = pickle.load(open("../call_MCMAC/SE_bootstrapRedshift.pickle", "r"))
NW_z = NW_z.flatten()
SE_z = SE_z.flatten()
```

```python
bins = np.arange(0.862, 0.876, 0.002)
vel_bins = [vrad(0.87, bin) for bin in bins]
vel_bins = np.append(np.array(vel_bins[:4]) * (-1), np.array(vel_bins[4:]))
vel_bins_labels = [str(int(bin)) for bin in vel_bins]
```

```python
paperpath = paperpath[0]
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.hist(NW_z, bins=50, normed=True, alpha=0.3)
ax1.hist(SE_z, bins=50, normed=True, alpha=0.3)
ax1.axvline(0.87, label=r"location of $z_{main}$", color='k', lw=3)
ax1.axvline(0.86842, color='b', ls='--', lw=3, label=r"location of $z_{NW}$")
ax1.axvline(0.87110, color='g', ls='--', lw=3, label=r"location of $z_{SE}$")
ax2 = ax1.twiny()

ax2.set_xticks(vel_bins)
ax2.set_xticklabels(vel_bins_labels)
ax1.set_ylabel("PDF")
ax1.set_xlabel("Bootstrapped redshift", fontsize=13)
ax2.set_xlabel("Radial velocity difference from z=0.87 (km / s)", fontsize=13)
ax1.legend(loc="best")
plt.savefig(paperpath + "bootstrapped_redshift.pdf", bbox_inches="tight")
print ""
```
