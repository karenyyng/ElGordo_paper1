# How to do the relic calculation 

## Info from Lindner et al. 2013

>"shock speed can be interpreted as an upper limit on the collision speed" 

implies that this a relative speed between the two subclusters? 

This is different than saying that the shock speed is in the CM frame.


### the position of the center of mass and thus the separations of the relics to the CM 

* using the full realizations takes up a lot of time for doing conversion between coordinates and calculating the physical separation since we have half a million realizations and the code is not vectorized - python loops are slow - this might be able to be solved by using num expression (numex) but I have not tried it yet

* using the mean location of the center of mass which is a lot faster but will not fold the uncertainties in --- maybe I should rewrite this part using Cython

### the velocities of the relics

we do not know the time evolution of the speed but we can assume an average velocity to be either: 

* the collisional velocity of the NW subcluster 

* the collision velocity of the NW subcluster times a slow-down factor



In the center of momentum frame, the speed of $m_1$ and $m_2$ are related by: 

$$v_{2, CM} = - \frac{m_1}{m_2} v_{1, CM}$$ -- (1) 

but in our simulation, we only know about the relative velocities of the two subclusters, 


$v_{rel} = v_1 - v_2 $ -- (2) which is true in any frame

Plug (1) in (2) to get 

$$ v_{1,CM} = \left( \frac{m_2}{m_1 + m_2} \right) v_{rel}$$ 
$$ v_{2,CM} = -\left( \frac{m_1}{m_1 + m_2} \right) v_{rel}$$


### Relic calculation - v.1 - just getting a feeling of how the distances compare 

* using a mean location for the center of mass 
* using NW $v_{relic} = 4300$ km/s $\pm 800 $ km/s assuming this is in CM frame 

Is this shock speed in 3D or 2D??? it makes a difference - should be in 3D...

Calculation of the simulated separation for relic from the CM: 

$$ s_{relic} = v_{relic_i} ~ t_{tsc_j} \cos(\alpha) $$

There will be 2 outputs from the calculations just from simulation since there are two TSCs. 


Unit conversion is done to convert units of (km /s * Gyr) to Mpc 

Mpc / km = $ 60 \times 60 \times 365 \times 24 \times 1e9 / (1e13 \times 1e6 \times 3.086) $

### Initialize the velocities 

perform the full on calculation including all the uncertainties

```python
data['v_relic'] = data['v_3d_col'] * data["m_1"] /\
    (data["m_1"] + data["m_2"]) 
```

```python
histplot1d(np.array(data["v_relic"]), prob=np.array(data["prob"]), 
           plot=True)
```

### Initialize the lower bound of the predicted location 

$$x_{NW} = \frac{m_{SE}}{m_{NW}} x_{SE} $$
$$x_{rel} = x_{NW} - x_{SE} $$
$$ x_{NW} = \frac{m_{SE}}{m_{NW} + m_{SE}} x_{rel}$$ 

```python
# convert m to Mpc , Myrs to seconds 
unitconversion = 60 * 60 * 365 * 24 * 1e9 / (1e13 * 1e6 * 3.086) 

data['r_proj0_min'] = data['d_3d'] * \
                        np.cos(data['alpha'] / 180. * np.pi) * \
                        (data['m_1']) / (data['m_2'] + data['m_1'])
data['r_proj1_min'] = data['d_3d'] * \
                        np.cos(data['alpha'] / 180. * np.pi) * \
                        (data['m_1']) / (data['m_2'] + data['m_1'])

data = data[~np.isnan(data['r_proj0_min'])]
data = data[~np.isnan(data['r_proj1_min'])]
```

### Initialize the upper bound of the predicted location

```python
data['r_proj0_max'] = (data['v_relic'] * data['TSM_0'] * 
                   np.cos(data['alpha'] / 180 * np.pi)) * unitconversion
data['r_proj1_max'] = (data['v_relic'] * data['TSM_1'] *
                   np.cos(data['alpha'] / 180 * np.pi)) * unitconversion
data = data[~np.isnan(data['r_proj0_max'])]
data = data[~np.isnan(data['r_proj1_max'])]
```
