```python
def N(x, mu, sigma):
    return 1/np.sqrt(2*np.pi)/sigma*np.exp(-np.power(x - mu, 2)/2./sigma**2)
```

We pick the std. as 10  to encompass the uncertainty
The most conservative model gives a projection angle of 90-51 = 39 degrees at 33% polarization 
the least conservative model gives a projection angle of 90 - 55 = 34 degrees at 33% polarization 
we can pick the std. dev. as 4 degrees then it 'll be the relative probability of the least conservative model to the most 
conservative model will be around 60%.

However, I want to be even more conservative than that so I pick a std. dev. as 10 degrees,
then the relative probability of the least conservative model to the most conservative model will be around 92%


```python
x = np.arange(0,90,.1)
plt.plot(x, N(x, 39, 4),'r-', label = r'$\sigma $= 4 deg ')
plt.plot(x, N(x, 39, 8),'b-', label = r'$\sigma $= 8 deg')
plt.plot(x, N(x, 39, 10),'g-', label = r'$\sigma $= 10 deg')
plt.legend(loc='best')
```
