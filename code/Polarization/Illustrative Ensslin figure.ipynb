# Want to illustrate how we come up with the 39 degree angle 

```python
upper_solid = pd.read_csv('upper_solid_curve_Ensslin98.csv', skiprows=5)
lower_solid = pd.read_csv('lower_solid_curve_Ensslin98.csv', skiprows=5)
upper_dash = pd.read_csv('upper_dashed_curve_Ensslin98.csv', skiprows=5)
lower_dash = pd.read_csv('lower_dashed_curve_Ensslin98.csv', skiprows=5)
```

```python
upper_solid["alpha"] = 90 - upper_solid["ViewingAngle(90-alpha)"]
upper_dash["alpha"] = 90 - upper_dash["ViewingAngle(90-alpha)"]
lower_solid["alpha"] = 90 - lower_solid["ViewingAngle(90-alpha)"]
lower_dash["alpha"] = 90 - lower_dash["ViewingAngle(90-alpha)"]
```

```python
upper_solid.columns
```

```python
plt.plot(upper_solid["Polarization%"], upper_solid["alpha"], 
         "r", label=r"strong field, $\alpha_{radio} = 1, \gamma=3, R=4$")
plt.plot(upper_dash["Polarization%"], upper_dash["alpha"], 
         "r--", alpha=0.6, label=r"weak field, $\alpha_{radio} = 1, \gamma=3, R=4$")
plt.plot(lower_solid["Polarization%"], lower_solid["alpha"], 
         "b", label=r"strong field, $\alpha_{radio} = 1.5, \gamma=4, R=2.5$")
plt.plot(lower_dash["Polarization%"], lower_dash["alpha"], 
         "b--", label=r"weak field, $\alpha_{radio} = 1.5, \gamma=4, R=2.5$")
plt.xlabel("Polarization %", size=12)
plt.ylabel(r"Projection angle $\alpha$", size=15)
plt.axvline(33, ymin=0, ymax=39./90, color="k", ls=":", lw=2)
plt.axhline(y=39, xmin=0, xmax=33./80, color="k", ls=":", lw=2)
plt.axvspan(32, 34, ymax=39./90, color="grey", alpha=0.3)
plt.minorticks_on()
plt.xlim(0, 80)
plt.legend(loc="upper right", fontsize=10)
plt.savefig("Ensslin_rep_polarization.pdf")
```

Polarization of synchrotron emission of shocked magnetic fields as a function of the viewing angle δ. The strong field case is given by solid lines, the weak field case by dashed lines. The upper pair of curves (solid & dashed) corresponds to a radio spectral index of α = 1, and therefore γ = 3 and R = 4. The lower lines correspond to α =1.5, γ = 4 and R =2.5. Observed polarizations of radio relics are plotted above the viewing angle predicted from the accretion shock theory (1253+257, 1712+64), simulations of a cluster merger (1706+78), or – if no data is available – at their best-fit position (0917+75). The large projected radius of 0917+75 indicates a large viewing angle, consistent with the best-fit position. Uncertainties in the angles are large

# Possible tighter constraint 

Lindner et al. 2013 section 4.5 (FINALIZED VERSION!)

$<\alpha> = 0.86 \pm 0.15, r = 2.74$, assumes $\gamma = 5/3$
r is the compression ratio according to the paper 

```python
p = np.arange(0.0, 1.005, 0.005)
alpha_mean = .86
g_mean = gamma(alpha_mean)
R_mean = compression_ratio(alpha_mean)
print g_mean
print "{0:.1f}".format(R_mean)

alpha_str = r"$\alpha_{radio}$"

plt.plot(p * 100, strong_field_alpha(g_mean, R_mean, p), 
         label=r"strong field, {0}={1}, $R=${2:.1f}, $\gamma$={3:.1f}".format(
         alpha_str, alpha_mean, R_mean, g_mean))
plt.plot(p * 100, weak_field_alpha(g_mean, R_mean, p), "b",
         label=r"weak field, {0}={1}, $R=${2:.1f}, $\gamma$={3:.1f}".format(
         alpha_str, alpha_mean, R_mean, g_mean), alpha=.5)

ymin, ymax = plt.ylim()
xmin, xmax = plt.xlim()
alpha = strong_field_alpha(g_mean, R_mean, 0.33)

plt.axvline(33, ymin=0, ymax=alpha / ymax, color="k", ls="--", lw=2)
plt.axhline(y=alpha, xmin=0, xmax=33./ xmax, color="k", ls=":", lw=2)
plt.axvspan(32, 34, ymax=alpha / ymax, color="grey", alpha=0.3)
plt.arrow(33. / 2, alpha, 0, -25, head_width=2, 
          head_length=2, fc='k', ec='k')
plt.legend(loc="upper right")
plt.xlabel("Polarization %", size=12)
plt.ylabel(r"$\alpha$", size=20)
plt.savefig("Ensslin_polar_fig.pdf", bbox_inches="tight")
plt.title("constraints if we use the mean radio emission spectral index")

print "projection angle wanted is {0:.2f}".format(alpha)
```
